```python
def findaverage(val1,val2,val3):
  #function findaverage will receieve 3 values and average the three values, finding the average to a third decimal point, printing it out
  avg = ("{:.3f}".format((val1+val2+val3)/3))
  print(f'The average is {avg}') 
```
